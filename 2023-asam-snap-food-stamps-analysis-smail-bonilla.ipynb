```python
## create a subset of data for only 2015, for later % change
df_old = df.query("year == 2015") 
df_old
```

```python
df_old.info()
```

```python
## fixed variable for all boroughs for SNAP, Asians alone

Bronx_old = df_old["households receiving food stamps and snap asian alone"].iat[0]
Brooklyn_old = df_old["households receiving food stamps and snap asian alone"].iat[1]
Manhattan_old = df_old["households receiving food stamps and snap asian alone"].iat[2]
Queens_old = df_old["households receiving food stamps and snap asian alone"].iat[3]
Staten_old = df_old["households receiving food stamps and snap asian alone"].iat[4]
NYC_old = df_old["households receiving food stamps and snap asian alone"].iat[5]
```

```python
## confirm values

print(f"Bronx_old: {Bronx_old}")
print(f"Brooklyn_old: {Brooklyn_old}")
print(f"Manhattan_old: {Manhattan_old}")
print(f"Queens_old: {Queens_old}")
print(f"Staten_old: {Staten_old}")
print(f"NYC_old: {NYC_old}")
```

# Calculating Bronx Percentage Change for SNAP

```python
df_bronx = df.query("borough == 'Bronx borough, Bronx County, New York'").copy()
df_bronx
```

```python
## Bronx percentage change for SNAP
df_bronx["Bronx_pctsnap"] = \
((df["households receiving food stamps and snap asian alone"] - Bronx_old) / Bronx_old) * 100

df_bronx
```
